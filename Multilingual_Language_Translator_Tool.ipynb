```python
# Install required libraries
!pip install googletrans==4.0.0-rc1 ipywidgets --quiet

# Import necessary modules
from googletrans import Translator, LANGUAGES
import ipywidgets as widgets
from IPython.display import display, HTML

# Initialize the Translator
translator = Translator()

# Text input widget
text_input = widgets.Textarea(
    value='Hello! How are you?',
    placeholder='Enter text to translate...',
    description='Text:',
    layout=widgets.Layout(width='100%', height='100px')
)

# Dropdown for source language
source_lang = widgets.Dropdown(
    options=[('Auto Detect', 'auto')] + [(name.title(), code) for code, name in LANGUAGES.items()],
    value='auto',
    description='From:'
)

# Dropdown for target language
target_lang = widgets.Dropdown(
    options=[(name.title(), code) for code, name in LANGUAGES.items()],
    value='hi',
    description='To:'
)

# Output area for result
output_area = widgets.Output()

# Translate button
translate_button = widgets.Button(description="Translate", button_style='success')

# Translation function
def translate_text(b):
    with output_area:
        output_area.clear_output()
        try:
            result = translator.translate(text_input.value, src=source_lang.value, dest=target_lang.value)
            print("Translated Text:\n")
            print(result.text)
        except Exception as e:
            print("Error:", e)

# Bind button click to function
translate_button.on_click(translate_text)

# Display UI components
display(HTML("<h3>Language Translation Tool</h3>"))
display(text_input, source_lang, target_lang, translate_button, output_area)
```
